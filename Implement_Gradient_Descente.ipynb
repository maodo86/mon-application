```python
import numpy as np
import matplotlib.pyplot as plt


#création de la courbe sin de x avec x entre -5:+5
fig, ax = plt.subplots()
x = np.linspace(-5,5,100)
f_x = 2*x**2
ax.plot(x, f_x, label="polynome", color='r')
ax.set_xlabel("poids")
ax.set_ylabel("impact")
ax.set_title("Analyse d'impact")
ax.legend()
plt.savefig("Courbe d'analyse.pdf")
plt.show()
```

```python
#Implémentation méthode Gradient Descente

#importation des bibliothèques
import numpy as np
import matplotlib.pyplot as plt

#domaine de définition de la fonction
x = np.arange(-10, 10, 1)
def f(x):
    return x**2-x+1
def grad_f(x):
    return 2*x-1
plt.plot(x, f(x))

#initialisation algo descente 
x = -9             #valeur initiale
alpha = 0.3        #pas de déplacement 
k = 0              #compteur des itérations
epsilon = 1e-5   #

#test d'arret: sur l'itération
while abs(grad_f(x)) > epsilon :
    plt.plot(x, f(x), marker ='o', color ='r')  #affichage de f(x) de x
    plt.plot(x, grad_f(x), marker ='o', color='r')
    x = x -alpha*grad_f(x)
    print("itération", k, "Fonction cout", f(x), "abscisse", x)
    k += 1
```
