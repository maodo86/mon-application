```python
import cv2 as cv
# Code optionnel pour Ã©liminer les warnings

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
from deepface import DeepFace
```

```python
chemin=r"C:\Users\dell\Pictures\photo_maodo.jpg" #"C:\Users\dell\Documents\Perso\happy boy.png"
img=cv.imread(chemin)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
predictions=DeepFace.analyze(img)
```

```python
predictions
```

```python
predictions['dominant_emotion']
```

```python
face_cascade=cv.CascadeClassifier(r"C:\Users\dell\Documents\Perso\haar-cascade-files-master\haarcascade_frontalface_default.xml")
```

```python
gray=cv.cvtColor(img, cv.COLOR_BGR2GRAY)
face_rect=face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=4)
for (x,y,w,h) in face_rect:
    cv.rectangle(img, (x,y),(x+w, y+h),(0,255,0), 2)
    cv.putText(img, str(predictions['dominant_emotion']), (20,20), cv.FONT_HERSHEY_COMPLEX, 1.0, (0,250,0), thickness=2)
```
