```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5,5,50)
def sigmoid_fc(x):
    y= 1/(1+np.exp(-x))
    plt.plot(x, y)
    plt.show()
    return y

    
```

```python
sigmoid_fc(x)
```

```python
def tanh_fc(x):
    y=(1-np.exp(-2*x))/(1+np.exp(-2*x))
    plt.plot(x, y)
    plt.show()
```

```python
tanh_fc(x)
```

```python
def RelU(x):
    if x<0:
        return 0
    else:
        return x
```

```python
RelU(0.5)
```

```python
def leaky_RelU(x, alpha=0.01):
    if x<0:
        return alpha*x
    else:
        return x
```

```python
leaky_RelU(-1)
```

```python
def swish(x,beta=0.01):
    y=2*x*sigmoid_fc(beta*x)
    plt.plot(x, y)
    plt.show()
```

```python
swish(x)
```

```python
#donne les probabilité de sortie de chaque classe
def softmax(x):   #généralisation de sigmoid, dernière couche pour la classification multiclass
    return np.exp(x)/np.exp(x).sum(axis=0)                 
```

```python
softmax([10, 2, 12])
```
